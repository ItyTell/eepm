```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
import scipy as scp
```

```python
t = sp.Symbol('t')
x = sp.Function('x')(t)
```

```python
N = 9
r = N / 3 
q = 2 * N / 3 
p = r * q / 4
equesion = sp.Eq(x.diff(t), r * x * (1 - x / q) - p)
```

### Стаціонарні точки

```python
x1 = sp.solve(equesion.rhs, x)[0]
```

```python
eq1 = sp.dsolve(equesion, x, ics={x.subs({t:0}): 2})
eq1
```

```python
eq2 = sp.dsolve(equesion, x, ics={x.subs({t:0}): 4})
eq2
```

```python
res1 = sp.lambdify((t), eq1.rhs)
res2 = sp.lambdify((t), eq2.rhs)
```

```python
T = np.linspace(0, 1.8, 100)
fig, ax = plt.subplots()
ax.plot(T, res1(T))[0].set_label("x(0) = 2")
ax.plot(T, res2(T))[0].set_label("x(0) = 4")
ax.plot(T, [x1] * len(T), 'b--')[0].set_label("стаціонарна точка")
ax.legend()
plt.show()
```

```python
r = 3
q = 6
eps = 1/8
p = r * q / 4 - eps * eps
print(f"p = {p}")
equesion = sp.Eq(x.diff(t), r * x * (1 - x / q) - p)
```

### Стаціонарні точки

```python
x1, x2 = sp.solve(equesion.rhs, x)
print(f"x1, x2 = {x1}, {x2}")
```

```python
eq1 = sp.dsolve(equesion, x, ics={x.subs({t:0}): 3})
eq1
```

```python
eq2 = sp.dsolve(equesion, x, ics={x.subs({t:0}): 4})
eq2
```

```python
eq3 = sp.dsolve(equesion, x, ics={x.subs({t:0}): 5})
eq3
```

```python
res1 = sp.lambdify((t), eq1.rhs)
res2 = sp.lambdify((t), sp.re(eq2.rhs))
res3 = sp.lambdify((t), eq3.rhs)
```

```python
T = np.linspace(0, 8, 100)
fig, ax = plt.subplots()
ax.plot(T, res1(T))[0].set_label("x(0) = 3")
ax.plot(T, res2(T))[0].set_label("x(0) = 4")
ax.plot(T, res3(T))[0].set_label("x(0) = 5")
ax.plot(T, [x1] * len(T), 'b--')[0].set_label("стаціонарна точка")
ax.plot(T, [x2] * len(T), 'r--')[0].set_label("стаціонарна точка")
ax.set_ylim(-5, 10)
ax.legend()
plt.show()
```
